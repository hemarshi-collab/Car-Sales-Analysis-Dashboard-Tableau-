PROBLEM STATEMENT

Salaries in the Data Science and AI job market vary widely based on experience, job role, location, company size, and remote work. However, it is not clear which factors influence salaries the most or how salaries have changed over time.

This project uses Exploratory Data Analysis (EDA) to analyze global salary data from 2020 to 2025 and uncover key patterns, trends, and insights to help job seekers and organizations make better data-driven decisions.

PROJECT OBJECTIVES

Understand salary distribution and trends

Analyze salary differences by experience, role, location, and company size

Study the impact of remote work on salaries

Identify high-paying roles and regions

Extract meaningful career and business insights

Key Variables:

work_year: Year of salary record

experience_level: Entry, Mid, Senior, Executive

employment_type: Full-time, Part-time, Contract, Freelance

job_title: Role designation

salary_in_usd: Annual salary standardized in USD

employee_residence: Country of employee

company_location: Country of employer

remote_ratio: Percentage of remote work (0, 50, 100)

company_size: Small, Medium, Large

1. The Goal
"The goal of this project was simple: I wanted to find out what actually controls your salary in the data world. Is it your job title? Is it where you live? Or is it working from home? I used a dataset of over 46,000 unique records to find the answer."

2. Data Cleaning
"I started with a large dataset, but it had many duplicates. I cleaned the data to make it accurate. I found that the average salary is around $146k, but the data is 'right-skewed.' This means while most people earn a good mid-range salary, a few experts earn very high amounts, which pulls the average up."

3. Key Findings (The "Big Three")
"Here is what the data told me:
Experience Matters Most: Senior and Executive roles don’t just pay more; their pay varies a lot. This means at higher levels, your specific skills give you more power to negotiate.
The Big Roles: Data Scientists and ML Engineers are the most common roles. If you want the highest pay, Specialized AI roles are the way to go.
Company Size: Large companies usually pay the best, but medium-sized startups often pay very high 'bonus' salaries to attract top talent."

4. Remote Work & Trends
"I also looked at remote work. Interestingly, fully remote jobs still compete well with office jobs in terms of pay. Looking at the trend from 2020 to 2025, salaries have been rising every year, especially after the AI boom in 2023."

5. Conclusion
In summary, my EDA shows that the Data/AI market is healthy and growing. To maximize salary, one should aim for Senior ML roles in Large or Medium companies. This project allowed me to apply my skills in Python and Data Viz to a real-world business problem.

```python
# What is the salary distribution
sns.histplot(df["salary_in_usd"], bins=50, kde=True, color="Red")
plt.title("Salary Distribution (USD)", fontsize=12, fontweight="bold")
plt.xlabel("Salary in USD", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

Salary distribution is right-skewed

Most salaries lie in the lower-to-mid range,
Most people earn moderate salaries, but a few earn very high salaries.

Few very high-paying roles increase the average salary

```python
# How many jobs per experience level
fig = px.histogram(
    df,
    x="experience_level",
    title="Experience Level Distribution",
    color="experience_level",
    text_auto=True)
fig.update_layout(
    xaxis_title="Experience Level",
    yaxis_title="Count",
    template="plotly_white",
    bargap=0.2,
    showlegend=False)
fig.show()
```

This chart shows the number of employees in each experience level category, helping us understand the overall experience distribution of the workforce.

I found that salary increases with experience.
Senior and executive roles earn significantly more than entry-level roles

```python
# Top 10 Jobs
top_jobs = df["job_title"].value_counts().head(10).reset_index()
top_jobs.columns = ["Job Title", "Number of Jobs"]
```

```python
# What are the most common job
fig = px.bar(
    top_jobs,
    x="Job Title",
    y="Number of Jobs",
    title="Top 10 Job Roles",
    text="Number of Jobs",
    color="Number of Jobs",
    color_continuous_scale="Blues")
fig.update_layout(
    xaxis_title="Job Title",
    yaxis_title="Number of Jobs",
    template="plotly_white",
    xaxis_tickangle=-45)
fig.show()
```

This interactive Plotly bar chart visualizes the top 10 job roles based on frequency, making it easy to compare job demand across roles.

```python
# How does salary vary by experience level
fig = px.box(
    df,
    x="experience_level",
    y="salary_in_usd",
    title="Salary vs Experience Level",
    points=False )
fig.update_layout(
    xaxis_title="Experience Level",
    yaxis_title="Salary in USD",
    template="plotly_white")
fig.show()
```

This interactive Plotly boxplot visualizes the distribution of salaries across experience levels, highlighting medians, ranges, and outliers with an interactive hover feature.

Entry-level employees earn less than executives.

Senior-level salaries vary a lot because of bonuses or special skills.

Some salaries are unusually high or low (outliers).

We can see salary outliers existed, I did not remove them because they represent real high-paying executive and specialized roles, which are important for market analysis.

```python
# Does employment type affect salary
fig = px.box(
    df,
    x="employment_type",
    y="salary_in_usd",
    title="Salary by Employment Type")
fig.update_layout(
    xaxis_title="Employment Type",
    yaxis_title="Salary in USD",
    template="simple_white")
fig.show()
```

Full-time roles usually have higher salaries.

Contract and part-time roles have more varied salaries

```python
# Does remote work affect salary
fig = px.box(
    df,
    x="remote_ratio",
    y="salary_in_usd",
    title="Salary vs Remote Work",
    points=False,
    color="remote_ratio")
fig.update_layout(
    xaxis_title="Remote Ratio",
    yaxis_title="Salary in USD",
    template="simple_white")
fig.show()
```

Full remote jobs can pay more or less than office jobs.

Part-time or partially remote jobs have more salary variation.

Shows unusual very high or low salaries.

```python
# Average salary by remote type
df.groupby("remote_ratio")["salary_in_usd"].mean()
```

```python
# Salary by experience & remote work
fig = px.box(
    df,
    x="experience_level",
    y="salary_in_usd",
    color="remote_ratio",
    title="Experience vs Salary by Remote Ratio",
    points=False)
fig.update_layout(
    xaxis_title="Experience Level",
    yaxis_title="Salary in USD",
    template="simple_white")
fig.show()
```

Higher experience usually means higher salary

Remote and non-remote roles may pay differently

Salary variation increases at higher experience levels

```python
# Salary by company size
fig = px.box(
    df,
    x="company_size",
    y="salary_in_usd",
    title="Salary by Company Size",
    points=False)
fig.update_layout(
    xaxis_title="Company Size",
    yaxis_title="Salary in USD",
    template="simple_white")
fig.show()
```

Large companies usually pay more

Small companies show wider salary variation

Some very high or low salaries stand out

```python
# Company size + experience (Heatmap)
pivot = df.pivot_table(
    values="salary_in_usd",
    index="experience_level",
    columns="company_size",
    aggfunc="mean").reset_index()
```

```python
fig = px.imshow(
    pivot.set_index("experience_level"),
    text_auto=".0f", color_continuous_scale= "Blues",
    title="Average Salary by Experience & Company Size")
fig.update_layout(
    xaxis_title="Company Size",
    yaxis_title="Experience Level")
fig.show()
```

This heatmap compares average salaries across experience levels and company sizes, making pay differences easy to spot at a glance.

Salaries increase with experience level

Large companies usually pay more than small ones

Senior roles in large companies earn the highest salaries

Entry-level roles in small companies earn the lowest salaries

```python
# Top-paying employee countries
df.groupby("employee_residence")["salary_in_usd"].mean().sort_values(ascending=False).head(10)
```

```python
# Top-paying company locations
top_countries = df.groupby("company_location")["salary_in_usd"].mean().sort_values(ascending=False).head(10).reset_index()

px.bar(top_countries,
    x="company_location",
    y="salary_in_usd",
    title="Top Paying Company Locations")
```

Quickly see which countries pay the highest salaries

Easy comparison of average salary differences

Useful for career planning and market analysis

```python
# How have salaries changed over time
yearly_salary = df.groupby("work_year")["salary_in_usd"].mean().reset_index()
```

```python
fig = px.line(
    yearly_salary,
    x="work_year",
    y="salary_in_usd",
title="Average Salary Trend (2020–2025)",markers=True)
fig.update_layout(
    xaxis_title="Year",
    yaxis_title="Average Salary in USD",
    template="simple_white")
fig.show()
```

This line chart shows the trend of average salaries over time, helping us understand how pay has changed across years.

Salaries generally increase year by year

Sudden jumps or drops show market changes

Useful for understanding industry growth

```python
# Which roles pay the most?
df.groupby("job_title")["salary_in_usd"].mean().sort_values(ascending=False).head(10)
```

```python
# Salary variability by experience level
df.groupby("experience_level")["salary_in_usd"].std()
```
